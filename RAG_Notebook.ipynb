## Part 1: Install Ollama

This section explains how to install Ollama on your Mac computer (M series chip)

```bash
#this command makes sure hombrew installed.
brew --version
```

```bash
#install ollama, the tool that will allow you to establish a RAG connection with open source LLMs
brew install ollama
```

```bash
#start ollama (if you have already installed ollama, you can instead use 'brew services restart ollama')
brew services start ollama
```

```bash
#pull and write the ollama manifest
ollama pull llama3
ollama pull nomic-embed-text
ollama pull llama2
ollama list - list available models
```

```bash
#Install Ollama Python package
pip install ollama
```
